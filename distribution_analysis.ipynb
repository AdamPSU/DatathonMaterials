```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt; plt.style.use('fivethirtyeight')

from distfit import distfit
```

# Visualize Continuous Target Distributions

### Load Continuous Data

```python
df = pd.read_csv('../data/podcast_time.csv')
y = np.array(df['Listening_Time_minutes'])
```

### Plot Continuous Target Distribution

```python
dfit = distfit()
dfit.fit(y)

dfit.plot(
    chart='pdf',
    n_top=1,
    title='Best Fit Distribution with Confidence Intervals',

    emp_properties={ 
        'color': '#222e2d',
        'linewidth': 2.5, 
        'linestyle': '-' 
    },

    pdf_properties={ 
        'color': '#000',
        'linewidth': 2.5, 
        'linestyle': '-' 
    },  

    bar_properties={ 
        'color': '#466160',
        'linewidth': 1, 
        'edgecolor': '#222e2d',
        'align': 'center' 
    },

    cii_properties={ 
        'color': '#000',
        'linewidth': 2.5, 
        'linestyle': 'dashed', 
        'marker': 'o',
        'size': 20,
        'color_sign_multipletest': '#73112f',
        'color_sign': '#73112f',
        'color_general': '#73112f'
    },

    fontsize=14,
    xlabel='Values',
    ylabel='Frequency'
)
```

```python
dfit.plot_summary()
```

# Visualize Discrete Target Distributions

### Load Discrete Data

```python
df = pd.read_csv("../data/loan_approval.csv")
y = np.array(df['loan_status'])
```

### Plot Discrete Target Distribution

```python
dfit = distfit(method='discrete')
dfit.fit_transform(y)
dfit.plot(
    n_top=1,
    title='Best Fit Distribution with Confidence Intervals',
    
    pdf_properties={ 
        'color': '#000',
        'linewidth': 2.5, 
        'linestyle': '-' 
    },
    
    bar_properties={ 
        'color': '#466160',
        'linewidth': 1, 
        'edgecolor': '#222e2d',
        'align': 'center' 
    },

    cii_properties={ 
        'color': '#000',
        'linewidth': 2.5, 
        'linestyle': 'dashed', 
        'marker': 'o',
        'size': 20,
        'color_sign_multipletest': '#73112f',
        'color_sign': '#73112f',
        'color_general': '#73112f'
    },
)
```
